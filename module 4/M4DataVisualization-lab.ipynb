## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
from random import randint

table_name = 'DatabaseDesireNextYear'  # the table you wish to describe


QUERY = """
select 
    DatabaseDesireNextYear, 
    total * 100.0 / (select count(*) from {}) percentage
from ( 
    select 
        DatabaseDesireNextYear,
        count(*) as total from {}
    group by DatabaseDesireNextYear
)
order by 
    percentage desc
limit 5
""".format(table_name, table_name)

df = pd.read_sql_query(QUERY,conn)

dbs = df['DatabaseDesireNextYear']
percents = df['percentage']

# Explode data
explode = (0.1, 0.0, 0.2, 0.3, 0.0) # list(map(lambda p: p / 150, percents))

# Colors
colors = ("orange", "cyan", "brown", "indigo", "beige") # ['#%06X' % randint(0, 0xFFFFFF) for _ in range(5)]

# Wedge properties
wp = {'linewidth': 1, 'edgecolor': "darkblue"}

# Autopct argument
autopct = lambda p: "{:.1f}%".format(p)

# Creating plot
fig, ax = plt.subplots(figsize=(10, 7))
wedges, texts, autotexts = ax.pie(percents,
                                  autopct=autopct,
                                  explode=explode,
                                  labels=dbs,
                                  shadow=True,
                                  colors=colors,
                                  startangle=90,
                                  wedgeprops=wp,
                                  textprops=dict(color="magenta"))

# Adding legend
ax.legend(wedges, dbs,
          title="Databases",
          loc="center left",
          bbox_to_anchor=(1.2, 0, 0.5, 1))

plt.setp(autotexts, size=10, weight="bold")
ax.set_title("Top 5 DB Respondent Wish to Learn Next Year")


plt.show()
```

Create a pie chart of the top 5 programming languages that respondents wish to learn next year. Label the pie chart with language names. Display percentages of each language on the pie chart.


```python
# your code goes here
from random import randint

table_name = 'LanguageDesireNextYear'  # the table you wish to describe


QUERY = """
select 
    LanguageDesireNextYear, 
    total * 100.0 / (select count(*) from {}) percentage
from ( 
    select 
        LanguageDesireNextYear,
        count(*) as total from {}
    group by LanguageDesireNextYear
)
order by 
    percentage desc
limit 5
""".format(table_name, table_name)

df = pd.read_sql_query(QUERY,conn)

langs = df['LanguageDesireNextYear']
percents = df['percentage']

# Explode data
explode = (0.1, 0.0, 0.2, 0.3, 0.0) # list(map(lambda p: p / 150, percents))

# Colors
colors = ['#%06X' % randint(0, 0xFFFFFF) for _ in range(5)]

# Wedge properties
wp = {'linewidth': 1, 'edgecolor': "darkblue"}

# Autopct argument
autopct = lambda p: "{:.1f}%".format(p)

# Creating plot
fig, ax = plt.subplots(figsize=(10, 7))
wedges, texts, autotexts = ax.pie(percents,
                                  autopct=autopct,
                                  explode=explode,
                                  labels=langs,
                                  shadow=True,
                                  colors=colors,
                                  startangle=90,
                                  wedgeprops=wp,
                                  textprops=dict(color="magenta"))

# Adding legend
ax.legend(wedges,
          langs,
          title="Programming Laguages",
          loc="center left",
          bbox_to_anchor=(1.2, 0, 0.5, 1))

plt.setp(autotexts, size=10, weight="bold")
ax.set_title("Top 5 Programming Languages Respondent Wish to Learn Next Year")


plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT
    Age,
    WorkWeekHrs, 
    CodeRevHrs 
FROM 
    {}
Where
    Age between 30 and 35

""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
df = df.groupby("Age").median()

df.plot(kind='bar', stacked=True)
plt.title('Median workweek hours and CodeRevHrs in the 30-35 age group')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT
    Age,
    ConvertedComp 
FROM 
    {}
Where
    Age between 45 and 60

""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
df = df.groupby("Age").median()

df.plot(kind='line')
plt.title('Median compensation distribution in the 45-60 age group')
plt.xlabel('Age')
plt.ylabel('Converted compensation')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT
    MainBranch,
    count(*) total
FROM 
    {}
group by
    1
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
df.set_index('MainBranch', inplace=True)

df.plot(kind='barh')
plt.xlabel('Count')
plt.title('Horizontal Bar Chart of MainBranch')
plt.show()
```

How many respondents indicated that they currently work with 'SQL'? 

```python
# your code goes here
def query_and_return_df(db):
    QUERY = """
    SELECT
        {},
        count(*) total
    FROM 
        {}
    group by
        1
    order by
        2
    """.format(db, db)

    return pd.read_sql_query(QUERY,conn)
    
db_worked_df = query_and_return_df('DatabaseWorkedWith')
lang_worked_df = query_and_return_df('LanguageWorkedWith')
dev_type = query_and_return_df('DevType')

ad_hoc_query = lambda table, db: pd.read_sql_query("""
select 
    {}, 
    sum(total) total
from (
    select 
        {}, 
        count(Respondent) total 
    from 
        {} 
    group by 
        Respondent having count({})=1 and {}='{}'
    )
group by 
    1
""".format(table, table, table, table, table, db), conn)



# db_worked_df.query("DatabaseWorkedWith =='MySQL'"), lang_worked_df.query("LanguageWorkedWith =='SQL'")


print("How many respondents indicated that they currently work with 'SQL'? ", lang_worked_df.query("LanguageWorkedWith =='SQL'")['total'])
print("How many respondents indicated that they work on 'MySQL' only? ", ad_hoc_query('DatabaseWorkedWith', 'MySQL')['total'])
print('Majority of the survey responders are? ', dev_type.value_counts())
```
